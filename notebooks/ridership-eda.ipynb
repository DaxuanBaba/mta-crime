## Brooklyn

```python
df_2023_brooklyn = df_2023_ridership[df_2023_ridership['borough'] == 'Brooklyn']
df_2023_brooklyn_top = df_2023_brooklyn.loc[df_2023_brooklyn.groupby(['tpransit_timestam'])['ridership'].idxmax()]
df_2023_brooklyn_top['station_complex'].value_counts()
```

```python
df_2023_brooklyn_top.head(10)
```

```python
c1 = df_2023_brooklyn['station_complex'] == 'Atlantic Av-Barclays Ctr (B,D,N,Q,R,2,3,4,5)'
c2 = df_2023_brooklyn['station_complex'] == 'Bedford Av (L)'

df_2023_brooklyn_stations = df_2023_brooklyn[c1 | c2]
```

```python
df_2023_day = df_2023_brooklyn_stations.groupby(['transit_timestamp','station_complex'])['ridership'].sum()
df_2023_day = pd.DataFrame(df_2023_day).reset_index()
df_2023_day['transit_timestamp'] = pd.to_datetime(df_2023_day['transit_timestamp'])
df_2023_day['weekday'] = df_2023_day['transit_timestamp'].dt.day_name()
df_2023_day.tail(10)
```
